```python
def area_of_rectangle(length:int, width:int=2)->int:
    area = length * width
    return area
```

```python
# function is called with both the inputs
area_of_rectangle(5, 8)
```

```python
# with one input given, default value of width (2) is used
area_of_rectangle(5)
```

```python
# Function with no fixed number of inputs
def area_of_rectangle(*args)->int:
    area = args[0] * args[1]
    return area
```

```python
area_of_rectangle(5, 8)
```

```python
# Function with key:value arguments
def area_of_rectangle(**kwargs)->int:
    area = kwargs["length"] * kwargs["width"]
    return area
```

```python
area_of_rectangle(length=5, width=8)
```

```python
class Animal:
    def __init__(self, num_of_legs, color):
        self.num_of_legs = num_of_legs
        self.color = color
    
    # Class method
    def make_sound(self, sound):
        print("This is a {0} animal and it {1}.".format(self.color, sound))
```

```python
# cat is an object of Animal class
cat = Animal(4, "gray")
cat.make_sound("mews")
```

```python
# elephant is an object of Animal class
elephant = Animal(4, "black")
elephant.make_sound("trumpets")
```

```python
# parent class
class Animal:
    def __init__(self, num_of_legs, color):
        self.num_of_legs = num_of_legs
        self.color = color
    
    def make_sound(self, sound):
        print("This is a {0} animal and it {1}.".format(self.color, sound))

# child class "Bird" inheriting parent class "Animal"
class Bird(Animal):
    def __init__(self, num_of_legs, color, can_fly=True):
        self.can_fly = can_fly 

        # calling the __init__ of the parent class
        Animal.__init__(self, num_of_legs, color)

    # overriding the parent class method
    def make_sound(self, sound):
        if self.can_fly == True:
            print("This {0} bird can fly and it {1}.".format(self.color, sound))
        else:
            print("This {0} bird can not fly and it {1}.".format(self.color, sound))
```

```python
duck = Bird(2, "white", True)
duck.make_sound("quacks")
```

```python
ostrich = Bird(2, "brown", False)
ostrich.make_sound("chirps")
```
